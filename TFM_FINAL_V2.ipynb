```python
import pandas as pd

def asignar_estacion(row):
    fecha = row['FechaSalida']
    lugar = row['Destino']

    if lugar == 'Europa':
        if pd.to_datetime('21-03-' + str(fecha.year)) <= fecha < pd.to_datetime('21-06-' + str(fecha.year)):
            return 'Primavera'
        elif pd.to_datetime('21-06-' + str(fecha.year)) <= fecha < pd.to_datetime('23-09-' + str(fecha.year)):
            return 'Verano'
        elif pd.to_datetime('23-09-' + str(fecha.year)) <= fecha < pd.to_datetime('21-12-' + str(fecha.year)):
            return 'Otoño'
        else:
            return 'Invierno'
    elif lugar == 'Centro y Norte America':
        if pd.to_datetime('20-03-' + str(fecha.year)) <= fecha < pd.to_datetime('21-06-' + str(fecha.year)):
            return 'Primavera'
        elif pd.to_datetime('21-06-' + str(fecha.year)) <= fecha < pd.to_datetime('22-09-' + str(fecha.year)):
            return 'Verano'
        elif pd.to_datetime('22-09-' + str(fecha.year)) <= fecha < pd.to_datetime('21-12-' + str(fecha.year)):
            return 'Otoño'
        else:
            return 'Invierno'
    elif lugar in ['Argentina', 'Chile', 'Perú', 'Brasil', 'Resto Sud America','Paraguay']:
        if pd.to_datetime('23-09-' + str(fecha.year)) <= fecha < pd.to_datetime('21-12-' + str(fecha.year)):
            return 'Primavera'
        elif pd.to_datetime('21-12-' + str(fecha.year)) <= fecha < pd.to_datetime('21-03-' + str(fecha.year + 1)):
            return 'Verano'
        elif pd.to_datetime('21-03-' + str(fecha.year + 1)) <= fecha < pd.to_datetime('21-06-' + str(fecha.year + 1)):
            return 'Otoño'
        else:
            return 'Invierno'
    elif lugar == 'Asia del Este y Pacifico':
        if pd.to_datetime('01-03-' + str(fecha.year)) <= fecha < pd.to_datetime('01-06-' + str(fecha.year)):
            return 'Primavera'
        elif pd.to_datetime('01-06-' + str(fecha.year)) <= fecha < pd.to_datetime('01-09-' + str(fecha.year)):
            return 'Verano'
        elif pd.to_datetime('01-09-' + str(fecha.year)) <= fecha < pd.to_datetime('01-12-' + str(fecha.year)):
            return 'Otoño'
        else:
            return 'Invierno'
    elif lugar == 'África del Sur':
        if pd.to_datetime('01-11-' + str(fecha.year)) <= fecha < pd.to_datetime('01-04-' + str(fecha.year + 1)):
            return 'Verano'
        else:
            return 'Invierno'

# Aplicar la funcion a todo el df para asignarle una estacion al año.
df['Estacion'] = df.apply(asignar_estacion, axis=1)
```

```python
import pandas as pd

df['FechaSalida'] = pd.to_datetime(df['FechaSalida'])

# Agregar el año como una columna adicional
df['Año'] = df['FechaSalida'].dt.year

# Agrupar por mes con año
df_por_mes = df.groupby([df['FechaSalida'].dt.year, df['FechaSalida'].dt.month]).sum()

# Agrupar por trimestre con año
df['Trimestre'] = df['FechaSalida'].dt.quarter
df_por_trimestre = df.groupby(['Año', 'Trimestre']).sum()

# Imprime los DataFrames resultantes
print("Agrupado por mes con año:")
print(df_por_mes)

print("\nAgrupado por trimestre con año:")
print(df_por_trimestre)
```

```python
# Definir los percentiles para clasificar los gastos
percentiles = df['GastoTotal'].quantile([0.25, 0.5, 0.75])

# Función para clasificar los gastos en 'alto', 'medio' o 'bajo'
def clasificar_gasto(gasto):
    if gasto > percentiles[0.75]:
        return 'Alto'
    elif gasto > percentiles[0.25]:
        return 'Medio'
    else:
        return 'Bajo'

# Aplicar la función a la columna 'GastoTotal' para obtener la clasificación
df['ClasificacionGasto'] = df['GastoTotal'].apply(clasificar_gasto)

# Imprimir el DataFrame con la nueva columna de clasificación de gastos
df
```
